```python
def batch_generator_like(batch_size = 2,num_epochs = 10,allow_smaller_final_batch = False):
    epoch_num = 0
    length = 3 # num_training_example
    idx = np.arange(length)
    while epoch_num < num_epochs:
        for i in range(0, length, batch_size):
            # i starts with 0, then batch_size, 2* batch_size
            batch_idx = idx[i: i + batch_size]
            if not allow_smaller_final_batch and len(batch_idx) != batch_size:
                # teminate the loop, when facing smaller batch size. then it won't yield the bloew code.
                break
            yield batch_idx
            
        epoch_num += 1
```

```python
generator = batch_generator_like(2 ,2)
```
