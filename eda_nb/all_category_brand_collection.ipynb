```python
s = time.time()
output = []
for file in ['main_cat_only.csv','brand_only.csv','main_category_only.csv']:
    f_path = '/data/ner_task/all_category/{}'.format(file)
    print ('f_path',f_path)
    df = pd.read_csv(f_path, error_bad_lines = False)
    output.append(df)
df = pd.merge(output[0],output[1], on = 'itemid', how = 'left').merge(output[2], on = 'itemid', how = 'left')
del output
gc.collect()
e = time.time()
print ((e-s)/60.0, 'min')
```

```python
s = time.time()
df.brand = df.brand.apply(lambda x: x.lower() if type(x) != type(1.0) else x)
df.brand = df.brand.str.replace(pat = "[^a-zA-Z0-9&-' ']", repl = "") # remove symbols except '&', remove numbers?
df = df[df.brand.str.len()>1] # remove single character
df.replace('',np.nan, inplace = True)
df.replace('-',np.nan, inplace = True)
df.replace('--',np.nan, inplace = True)
df.replace('.',np.nan, inplace = True)
df.replace('0',np.nan, inplace = True)
df.replace('none',np.nan, inplace = True)
df.replace('tidak ada merek',np.nan, inplace = True)
df.replace('tidak ada merk',np.nan, inplace = True)
df.replace('no brand',np.nan, inplace = True)
df.dropna(subset = ['brand'], inplace = True)
df.dropna(subset = ['main_cat','main_category'], inplace = True)
print ('number of main_category : {}'.format(df.main_category.nunique()))
print ('main_category : {}'.format(df.main_category.unique().tolist()))

e = time.time()
print ((e-s)/60.0, 'min')
```
